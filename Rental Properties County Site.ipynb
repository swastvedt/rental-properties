```python
import requests
import re
import pandas as pd
from bs4 import BeautifulSoup
from bs4 import SoupStrainer

url = 'http://www16.co.hennepin.mn.us/pins/pidsrch.jsp'
url2 = 'http://www16.co.hennepin.mn.us/pins/pidresult.jsp'

#Load data with property apn numbers
licenses = pd.read_csv('Active_Rental_Licenses.csv', dtype = {'apn': str})
licenses.set_index('apn', drop=True, inplace=True)

#Input property apn number and retrieve owner, sale date, sale price and estimated market value
def requestInfo(number):
    data = {
        'pid' : str(number),
    }
    owner = 'empty'
    salePrice = 'empty'
    saleDate = 'empty'
    marketValue = 'empty'
    cellList = []
    
    only_td_tags = SoupStrainer('td')

    #Set up a session to store browser information
    s = requests.session()
    #Submit form with property id
    r = s.post(url2, data = data)

    #Parse the html result, selecting only table cells
    soup = BeautifulSoup(r.text, "lxml", parse_only=only_td_tags)

    ##Find cells adjacent to cells labled owner, sale date, sale price and estimated market value
    ##Save values in those cells as unicode, stripping leading and trailing whitespace
    tags = soup.find_all('p')
    for tag in tags:
        
        if tag.string == 'Owner \r\n      Name:':
            owner = unicode(tag.parent.next_sibling.next_sibling.string).rstrip().lstrip()
            
        if tag.find('span') != None:
            if tag.find('span').string != None and 'Sale Date:' in tag.find('span').string:
                saleDate = unicode(tag.parent.next_sibling.next_sibling.string).rstrip().lstrip()

            if tag.find('span').string != None and 'Sale Price:' in tag.find('span').string:
                salePrice = unicode(tag.parent.next_sibling.next_sibling.find('p').string).rstrip().lstrip()

            if tag.find('span').string != None and 'Estimated Market Value:' in tag.find('span').string:
                marketValue = unicode(tag.parent.next_sibling.next_sibling.find('p').string).rstrip().lstrip()
    
    print [owner, saleDate, salePrice, marketValue]
    return [owner, saleDate, salePrice, marketValue]
```

```python
#Test function on a single property apn
requestInfo('3602924220021')
```

```python
#Look up information for all properties in the licenses table
df2 = pd.DataFrame(columns=['apn','owner','saleDate','salePrice','marketValue'])

for apn in licenses.index:
    newRow = requestInfo(apn)
    df2 = df2.append({'apn' : apn, 'owner' : newRow[0], 'saleDate' : newRow[1], 'salePrice' : newRow[2], 'marketValue' : newRow[3]}, ignore_index=True)

df2
```

```python
df2.to_csv('PropertyInfo.csv')
```

```python
#Load the resulting table and format it so it can be joined to the original licenses table
propertyInfo = pd.read_csv('PropertyInfo.csv', dtype = {'apn': str}, usecols=[1,2,3,4,5])
propertyInfo.set_index('apn', drop=True, inplace=True)
propertyInfo['apnCopy'] = propertyInfo.index

propertyInfo.drop_duplicates(inplace=True)
propertyInfo.drop('apnCopy', axis=1, inplace=True)

propertyInfo.info()
```

```python
propertyInfo.index = propertyInfo.index.map(str)
```

```python
licenses.index = licenses.index.map(str)
```

```python
#Join licenses and property info tables
fullTable = licenses.join(propertyInfo)
fullTable.info()
```
